```python
import sys


def form_matrix(raw_matrix, size):
    final = []
    rows = size[0]
    columns = size[1]
    for i in range(rows):
        final.append(raw_matrix[columns * i: columns * (i + 1)])
    return final


def extract_input():
    # raw_data = sys.stdin.readline().split(" ")
    nested_size = []
    nested_matrix = []
    for line in sys.stdin.readlines():
        raw_data = line.split()
        nested_matrix.append([float(x) for x in raw_data[2:]])
        nested_size.append([int(x) for x in raw_data[0:2]])
    return nested_matrix, nested_size


def multiply_matrix(left, right):
    # make sure they have good dimensions, but I don't know how to throw an exception hihi
    try:
        rows_left, common, columns_right = len(left), len(left[0]), len(right[1])
        result = []
        for i in range(rows_left):
            temp = []
            for j in range(columns_right):
                element = 0
                for k in range(common):
                    element += left[i][k] * right[k][j]
                temp.append(round(element, 2))
            result.append(temp)
        return result
    except:
        return "There is something wrong"


def estimate_emission(transition_matrix, emission_matrix, initial_state):
    return multiply_matrix(multiply_matrix(initial_state, transition_matrix), emission_matrix)


def array_to_string(array):
    result = ""
    for i in range(len(array) - 1):
        result += str(array[i]) + " "
    result += str(array[len(array)-1])
    return result


def main():
    nested_matrix, nested_size = extract_input()

    transition_line, transition_size = nested_matrix[0], nested_size[0]
    transition_matrix = form_matrix(transition_line, transition_size)

    emission_line, emission_size = nested_matrix[1], nested_size[1]
    emission_matrix = form_matrix(emission_line, emission_size)

    initial_line, initial_size = nested_matrix[2], nested_size[2]
    initial_matrix = form_matrix(initial_line, initial_size)

    result = estimate_emission(transition_matrix, emission_matrix, initial_matrix)

    final = initial_size
    final[1] = len(emission_matrix[0])
    final += result[0]

    kattis_result = array_to_string(final)
    sys.stdout.write(kattis_result)
    
main()
```
