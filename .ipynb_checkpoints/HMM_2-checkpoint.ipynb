```python

import sys

def extract_input():
    # raw_data = sys.stdin.readline().split(" ")
    nested_size = []
    nested_matrix = []
    nested_emissions_count = 0
    nested_emissions = []
    i = 0
    for line in sys.stdin.readlines():
    # for line in input:
        i += 1
        raw_data = line.split()
        if (i == 4):
            nested_emissions = [int(x) for x in raw_data[1:]]
            nested_emissions_count = int(raw_data[0])
        else:
            nested_matrix.append([float(x) for x in raw_data[2:]])
            nested_size.append([int(x) for x in raw_data[0:2]])
    return nested_matrix, nested_size, nested_emissions, nested_emissions_count


def form_matrix(raw_matrix, size):
    final = []
    rows = size[0]
    columns = size[1]
    for i in range(rows):
        final.append(raw_matrix[columns * i: columns * (i + 1)])
    return final

def array_to_string(array):
    result = ""
    for i in range(len(array) - 1):
        result += str(array[i]) + " "
    result += str(array[len(array)-1])
    return result

def reverse_list(list):
    index = len(list) - 1
    reversed = []
    while index >= 0:
        reversed.append((list[index]))
        index -= 1
    return reversed

def compute_most_probable_sequence(A, B, Pi, O, T):
    delta_matrix, indices_matrix = compute_detla_matrix(A, B, Pi, O, T)
    N = len(A)
    index = 0
    prob = 0

    # find the last most probable state
    for j in range(N):
        if (delta_matrix[T-1][j] > prob):
            index = j
            prob = delta_matrix[T-1][j]

    # find the list of states using the indices_matrix
    indices = []
    indices.append(index)
    for t in range(T-1):
        indices.append(indices_matrix[T - t - 1][index])
        index = indices_matrix[T - t - 1][index]

    return array_to_string(reverse_list(indices))


def compute_detla_matrix(A, B, Pi, O, T):
    rows = T  # T (number of Observations)
    cols = len(A)  # N (number of states)
    delta_matrix = [[0 for _ in range(cols)] for _ in range(rows)]  # matrix T * N
    indices_matrix = [[-1 for _ in range(cols)] for _ in range(rows)]  # matrix T * N where we store the indices

    # fill in the initial states
    for i in range(cols):
        delta_matrix[0][i] = B[i][O[0]] * Pi[0][i]

    # fill the other states
    for t in range(1, rows):
        for i in range(cols):
            for j in range(cols):
                temp = A[j][i] * delta_matrix[t - 1][j] * B[i][O[t]]
                if (temp > delta_matrix[t][i]):
                    delta_matrix[t][i] = temp
                    indices_matrix[t][i] = j

    return delta_matrix, indices_matrix


def main():
    nested_matrix, nested_size, nested_emissions, nested_emissions_count = extract_input()
    A_raw, B_raw, Pi_raw = nested_matrix[0], nested_matrix[1], nested_matrix[2]
    A_size, B_size, Pi_size = nested_size[0], nested_size[1], nested_size[2]

    A, B, Pi = form_matrix(A_raw, A_size), form_matrix(B_raw, B_size), form_matrix(Pi_raw, Pi_size)

    O = nested_emissions
    T = nested_emissions_count

    sys.stdout.write(compute_most_probable_sequence(A, B, Pi, O, T))

main()

```
